```python
# Install required packages if not already installed
!pip install streamlit pyngrok 

import os
from pyngrok import ngrok

# Create the Streamlit app file
with open('app.py', 'w') as f:
    f.write("""
import streamlit as st
import tensorflow as tf
import numpy as np
import pickle
import re
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model

# Page configuration
st.set_page_config(
    page_title="Medical Diagnosis Assistant",
    page_icon="💊",
    layout="centered",
    initial_sidebar_state="collapsed"
)

# Custom CSS
st.markdown(\"\"\"
<style>
    .main-header {
        font-size: 2.5rem;
        color: #ff1493;
        text-align: center;
        margin-bottom: 1rem;
    }
    .sub-header {
        font-size: 1.2rem;
        color: #ff69b4;
        text-align: center;
        margin-bottom: 2rem;
    }
    .diagnosis-header {
        font-size: 1.5rem;
        font-weight: bold;
        margin-top: 1rem;
        margin-bottom: 0.5rem;
    }
    .diagnosis-item {
        background-color: #f8f9fa;
        padding: 1rem;
        border-radius: 0.5rem;
        margin-bottom: 0.5rem;
    }
    .probability-bar {
        height: 0.5rem;
        background-color: #ff69b4;
        margin-top: 0.5rem;
    }
    .disclaimer {
        font-size: 0.8rem;
        color: #6c757d;
        text-align: center;
        margin-top: 2rem;
        padding: 1rem;
        background-color: #f8f9fa;
        border-radius: 0.5rem;
    }
</style>
\"\"\", unsafe_allow_html=True)

# Headers
st.markdown('<div class="main-header">✨ Medical Diagnosis Assistant ✨</div>', unsafe_allow_html=True)
st.markdown('<div class="sub-header">Describe your symptoms and get instant disease predictions</div>', unsafe_allow_html=True)

# Session state init
for key in ['model', 'tokenizer', 'vectorizer', 'label_encoder']:
    if key not in st.session_state:
        st.session_state[key] = None
if 'max_len' not in st.session_state:
    st.session_state.max_len = 100

# Load model and components
@st.cache_resource
def load_model_and_components():
    try:
        model = load_model('/kaggle/input/lamia/tensorflow2/default/1/final_advanced_model.h5') 
        with open('/kaggle/input/lamia/tensorflow2/default/1/tokenizer.pickle', 'rb') as f:
            tokenizer = pickle.load(f)
        with open('/kaggle/input/lamia/tensorflow2/default/1/vectorizer.pickle', 'rb') as f:
            vectorizer = pickle.load(f)
        with open('/kaggle/input/lamia/tensorflow2/default/1/label_encoder.pkl', 'rb') as f:
            label_encoder = pickle.load(f)
        return model, tokenizer, vectorizer, label_encoder
    except Exception as e:
        st.error(f"Error loading model: {e}")
        return None, None, None, None

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    return text

def prepare_inputs(text):
    text = preprocess_text(text)
    bow = st.session_state.vectorizer.transform([text]).toarray()
    seq = pad_sequences(st.session_state.tokenizer.texts_to_sequences([text]), maxlen=20)
    return [bow, seq]

def predict_diagnosis(text):
    inputs = prepare_inputs(text)
    prediction = st.session_state.model.predict(inputs)
    top_idx = np.argsort(prediction[0])[-3:][::-1]
    top_probs = prediction[0][top_idx]
    top_diseases = st.session_state.label_encoder.inverse_transform(top_idx)
    return list(zip(top_diseases, top_probs))

# Load model
with st.spinner("Loading model..."):
    model, tokenizer, vectorizer, label_encoder = load_model_and_components()
    st.session_state.model = model
    st.session_state.tokenizer = tokenizer
    st.session_state.vectorizer = vectorizer
    st.session_state.label_encoder = label_encoder
    model_loaded = all([model, tokenizer, vectorizer, label_encoder])

if model_loaded:
    st.success("Model loaded successfully!")
else:
    st.error("Failed to load model.")

# User input
symptoms_input = st.text_area("Enter your symptoms here...", height=100, help="Include duration, intensity, etc.")

# Prediction
if st.button("Get Diagnosis", type="primary"):
    if not symptoms_input:
        st.warning("Please enter your symptoms.")
    elif model_loaded:
        with st.spinner("Analyzing..."):
            results = predict_diagnosis(symptoms_input)
            st.markdown('<div class="diagnosis-header">Top 3 possible diagnoses:</div>', unsafe_allow_html=True)
            for i, (disease, prob) in enumerate(results, 1):
                st.markdown(
                        "<div class='diagnosis-item'>"
                        f"<strong>Disease {i}:</strong> {disease} <br><strong>Probability:</strong> {prob*100:.1f}%"
                        f"<div class='probability-bar' style='width: {prob*100}%;'></div>"
                        "</div>",
                        unsafe_allow_html=True
                    )

            st.markdown("### Recommendations")
            st.info("This is not a replacement for professional medical advice.")

# Disclaimer
st.markdown(\"\"\"
<div class="disclaimer">
    <strong>Disclaimer:</strong> This tool is for educational use only and does not replace medical consultation.
</div>
\"\"\", unsafe_allow_html=True)


# Sidebar
with st.sidebar:
    st.header("About")
    st.write("This ML-powered tool predicts possible medical conditions based on symptom descriptions.")
    st.header("How to use")
    st.write("Type symptoms, be specific and include duration and severity")


""")

# Start ngrok and Streamlit app
if __name__ == "__main__":
    ngrok.set_auth_token("2wjFhuvDDwLNzSv3hMLByWp2Lf2_6XPP5MEjwfKLvAKzu2VQa")  # Replace with your actual token
    public_url = ngrok.connect(8501)
    print(f"Streamlit is accessible at: {public_url}")
    os.system("streamlit run app.py")
```
